### STUFF DOCUMENT CHAIN TEXT SUMMARIZATION !  

```python
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader("apjspeech.pdf")
docs = loader.load_and_split()
docs
```

```python

## NO WE SUMMARIZE IT USING THE STUFF DOCUMENTS CHAIN 
## First create a Prompt Templates

template =  """ Write a concise and short summary of the following speech,
Speech :{text}

"""
prompt=PromptTemplate(input_variables=['text'],
                      template=template)
```

```python
from langchain.chains.summarize import load_summarize_chain
## Load Summarize Chain for  whenever we want to create a text summarize we need to 
## import the summarize chain library for combine the  entire thing wwhich we are implemention the text summarization !  
```

```python
## Creating a Chain 

chain = load_summarize_chain(llm,chain_type='stuff', prompt = prompt, verbose = True)


## load_summarize_chain ?
## we have list od doc and it will combine all docu. and create a chain and then it will to llm model so that we get the entire text summarize of that docu.

output_summary = chain.run(docs)
output_summary
```

### Map reduce to Summarize Large documents

```python
## Similarly how map reduce work 

# 1)  ALL DOCUMENT CHUNK WILL BE MADE !

# 2) EACH CHUNK TRY TO CREATE A PROMPT TEMPLATES AND THAT PROMPT CAN BE SAME !

# 3) GOING TO SUMMARIZE THE CHUNK AND THEN WILL ABLE TO GET THE OUTPUT !  

# 4) WHEN COMBINNE ALL CHUNK AND SEND IT WITH ANOTHER PROMPT TEMPLATES ! 

# 5) FINALLY TO GET THE OUTPUT ! 

 
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
docs
```

```python
## use text splitter for creating object 
## Between the document so that it fit and didnt overlap 
## Divide the docs

final_documents=RecursiveCharacterTextSplitter(chunk_size=2000,chunk_overlap=100).split_documents(docs)
final_documents

```

```python
# WHT HAPPEND IF WE USE MAP REDUCE ! 

#  TAKE THE CHUNK ADN PASS TO THE PROMPT TEMPLATES THEN IT WILL SUMMARIZE IT LLM  MODELS   
# SIMILARLY IT SEND TO NEXT SENTENCE AND THEN TO THE PROMPT TEMPLATE ALONG WITHT THE  LLM MODEL AND THEN SUMMRIZE IT ! 
# SUMMARIZE IN THE SEPARATE WAY ! 
# THEN WE CAN COMBINE THEM WITH ANOTHER  PROMPT TEMPLATS TO GET THE FINAL SUMMARY ! 
# 

len(final_documents)
```

```python
prompt 
```

```python
## create a prompt ! FOR  THIS FOR THE CHUNKS ! 

chunks_prompt="""
Please summarize the below speech:
Speech:`{text}'
Summary:
"""
map_prompt_template=PromptTemplate(input_variables=['text'],
                                    template=chunks_prompt)
```

```python
## FINAL PROMPT TEMPLATES  

final_prompt='''
Provide the final summary of the entire speech with these important points.
Add a Motivation Title,Start the precise summary with an introduction and provide the summary in number 
points for the speech.
Speech:{text}

'''
final_prompt_template=PromptTemplate(input_variables=['text'],template=final_prompt)
final_prompt_template
```

```python
summary_chain=load_summarize_chain(
    llm=llm,
    chain_type="map_reduce",
    map_prompt=map_prompt_template,
    combine_prompt=final_prompt_template,
    verbose=True
)

output=summary_chain.run(final_documents)
output
```

### Refine Chain For Summarization

```python
chain=load_summarize_chain(
    llm=llm,
    chain_type="refine",
    verbose=True
)
output_summary=chain.run(final_documents)
output_summary
```
